# **Script para Colocar os Resultados e os Pesos no Drive**

```python
import shutil
import os

source_folder = '/content/yolov5/runs/detect/exp5'
destination_folder = '/content/drive/MyDrive/TG/arquiteturas/Ultralytics/YOLOv5/Resultados/Teste'

# Create the destination folder if it doesn't exist
if not os.path.exists(destination_folder):
    os.makedirs(destination_folder)

# Iterate through all files in the source folder
for file_name in os.listdir(source_folder):
    source_file = os.path.join(source_folder, file_name)
    destination_file = os.path.join(destination_folder, file_name)

    # Check if it's a file (and not a folder)
    if os.path.isfile(source_file):
        # Copy the file to the destination folder
        shutil.copy2(source_file, destination_file)

print("All files have been copied successfully.")
```
